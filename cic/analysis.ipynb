```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
wave = pd.read_csv("wave.csv")
wave
```

```python
plt.magnitude_spectrum(wave['i_data'], color ='green', Fs=10e6, scale='dB') 
plt.show()
```

```python
plt.magnitude_spectrum(wave['o_data'].loc[wave['o_ready'] == 1], color ='green', Fs=10e6/5, scale='dB') 
plt.show()
```

```python
wave['comb'] = wave['o_ready'] - wave['o_ready'].shift(+3)
wave = wave.dropna()
plt.magnitude_spectrum(wave['comb'], color ='green', Fs=10e6, scale='dB') 
plt.show()
```
